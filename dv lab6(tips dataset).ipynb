```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("tips.csv")
df
```

```python
a=pd.DataFrame(df['day'].value_counts())
print(a)
a.reset_index(inplace=True)
print(a)
plt.bar(a['index'],a['day'])
```

```python
sns.barplot(x='index',y='day',data=a)
plt.title("day vs index")
plt.show()
```

```python
plt.scatter(df['total_bill'],df['tip'])
plt.title("Bills vs Tips")
plt.xlabel("Bills")
plt.ylabel("Tips")
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df,hue='day')
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True)
plt.figure(figsize=(5,5))
plt.show()
```

```python
sns.pairplot(data=df,hue='sex')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
df.groupby('day').mean()['total_bill'].plot(kind='line')
plt.ylabel('mean')
plt.title("day vs mean")
plt.show()
```

```python
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.title("Total bill Distribution")
plt.xlabel("bill")
plt.ylabel("frequency")
plt.show()
```

```python
d={'tb':[16.99,10.34,21.01,23.68,24.59],
  'tip':[1.01,1.66,3.5,3.31,3.61]}
df1=pd.DataFrame(d)
plt.errorbar(range(len(df1)),df1['tb'],yerr=df1['tip'],fmt='o',color='g',label='tb with error')
plt.xticks(range(len(df1)),range(1,len(df1)+1))
plt.legend()
plt.xlabel('Datapoint Index')
plt.ylabel('Total bill')
plt.title("Error bar chart for dashboard")
plt.grid(True)
plt.show()
```
