```python
data = [30, 25, 15, 10, 20]
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
colors = ['red', 'green', 'orange', 'purple','skyblue']
plt.pie(data, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Pie Chart Example')
plt.show()
```

```python
import seaborn as sn
k=sn.load_dataset("iris")
g=sn.FacetGrid(k,col='species')
g.map(plt.scatter, "sepal_length", "sepal_width")
g.set_titles("{col_name}")
plt.show()
```

```python
d=sn.pairplot(k,hue='species',markers=['o','D','s'])
plt.show()
```

```python
x_data = [1, 2, 3, 4, 5]
y_data = [3, 4, 6, 8, 10]
y_data1 = [1, 2, 3, 4, 5]
plt.stackplot(x_data, y_data, y_data1, labels=['Series 1', 'Series 2'], alpha=1.0)
plt.title('Stacked Area Chart Example')
plt.legend()
plt.show()
```

```python
data = sn.load_dataset("iris")
sn.violinplot(x="species", y="sepal_length", data=data, palette="muted")
plt.xlabel('Species')
plt.ylabel('Sepal Length') 
plt.title('Violin Plot Example')
plt.show()
```
