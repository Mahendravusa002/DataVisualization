```python
# importing the necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
hea_ds=pd.read_csv("health.csv")
hea_ds.head()
```

```python
hea_ds=hea_ds.dropna(axis=1)
```

```python
import numpy as np

hea_ds['patient_status']=np.array(['Good' if i>=0.5 else 'Bad' for i in hea_ds['Health_Score'] ])
hea_ds
```

```python
for i in hea_ds:
    k=sum(hea_ds[i].isnull())
    print(i,":",k)
```

```python
import seaborn as sns
import plotly.express as px
# Tree map for the 
ax=px.treemap(hea_ds,path=['Health_Camp_ID'],
              title='Camp Wise Patients')
ax.show()
```

```python
camp_patient_count = hea_ds.groupby('Health_Camp_ID')['Patient_ID'].count().reset_index()
# Create a bar chart
fig = px.bar(hea_ds, 
             x='Health_Camp_ID', 
             y='Patient_ID', 
             title='Number of Patients from Each Health Camp',
             labels={'Health_Camp_ID': 'Health Camp ID', 'Patient_ID': 'Number of Patients'},
            color_discrete_sequence=['blue'])
fig.show()
```

```python
total_patients = hea_ds['Patient_ID'].nunique()
total_health_camps = hea_ds['Health_Camp_ID'].nunique()
# Creating a bar chart
data = {'Category': ['Total Patients', 'Total Health Camps'],
        'Count': [total_patients, total_health_camps]}
fig = px.bar(data, x='Category', y='Count', text='Count', title='Total Patients and Health Camps')
# Show the bar chart
fig.show()
```

```python
fig = px.pie(hea_ds, names='Donation', title='Distribution of Donations from Patients')
# Show the pie chart
fig.show()
```

```python
hea_ds['Discahrge']=np.array([1 if i=='Good' else 0 for i in hea_ds['patient_status'] ])
hea_ds
```

```python
fig = px.box(hea_ds, x='patient_status', y='Health_Score', title='Distribution of Health Scores')
fig.show()
```

```python
fig = px.scatter(hea_ds, x='Donation', y='Health_Score', color='patient_status', title='Donation vs. Health Score')
fig.show()
```

```python
fig = px.line(hea_ds, x='Health_Camp_ID', y='Health_Score', color='patient_status', 
              title='Health Score Trends Over Health Camps')
fig.show()
```

```python
patient_status_count = hea_ds['patient_status'].value_counts().reset_index()
# Creating a bar chart
fig = px.bar(patient_status_count, x='index', y='patient_status', text='patient_status', 
             title='Number of Patients Categorized as Good and Bad',
             labels={'index': 'Patient Status', 'patient_status': 'Number of Patients'})
# Show the bar chart
fig.show()
```
