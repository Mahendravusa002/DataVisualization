```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.show([0,6,0,20])
plt.show()
```

```python
year=[1972,1982,1992,2002,2012]
e_ind=[100.6,158.61,305.54,394.96,724.79]
e_ban=[10.5,25.21,58.65,119.27,274.87]
plt.plot(year,e_ind,color='orange',label='india')
plt.plot(year,e_ban,color='g',label='bangladesh')
plt.xlabel('years')
plt.ylabel('power consumption in kw')
plt.title('Electricity consumption if India and Bangladesh')
plt.legend()
plt.show();
```

```python
x=y=np.arange(-5,6,1)
x,y=np.meshgrid(x,y)
z=x**2+y**2
plt.imshow(z,cmap='plasma',extent=(min(x[0]),max(x[0]),min(y[:,0]),max(y[:,0])))
plt.colorbar(label='z')
plt.xlabel('x')
plt.ylabel('y')
plt.title('2d heatmap')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
x=y=np.arange(-5,6,1)
x,y=np.meshgrid(x,y)
z=x**2+y**2
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(x,y,z,cmap='plasma')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3d heatmap')
plt.show()
```
