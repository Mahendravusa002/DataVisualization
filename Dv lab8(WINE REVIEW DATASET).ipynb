```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
wr= pd.read_csv('winemag-data_first150k.csv')
wr.head()
```

```python
#Scatter plot
plt.figure(figsize=(10,6))
sns.scatterplot(data=wr,x='points',y='price')
plt.title('Scatter plot for points and price')
plt.show()
```

```python
#histogram
plt.figure(figsize=(10,6))
sns.histplot(data=wr,x='points',bins=20,kde=True)
plt.title('Histogram for points')
plt.show()
```

```python
#Bar chart
plt.figure(figsize=(12,6))
sns.countplot(data=wr,x='country',order=wr['country'].value_counts().head(10).index)
plt.title('Top 10 countries with most wine Reviews')
plt.xticks(rotation=45)
plt.show()
```

```python
#Heat map
plt.figure(figsize=(10,8))
hm=wr.corr()
sns.heatmap(hm,annot=True,cmap='coolwarm')
plt.title('Correlatation Heatmap')
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(10,10))
sns.boxplot(data=wr,x='points',y='country')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```
