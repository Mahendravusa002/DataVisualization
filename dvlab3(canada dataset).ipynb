```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
df=pd.read_csv("canadian_immigrationdata.csv")
print(df.columns)
df
```

```python
k=df.sort_values(['Total'],ascending=False)
m=k.head(6)
farea=m.iloc[:,4:-1].T
colm=m['Country']
farea.columns=colm
farea.plot.area()
```

```python
xc=df.columns[4:-1]
yc=df.iloc[182,4:-1]
fig=plt.figure(figsize=(30,4))
plt.bar(xc,yc)
plt.title('Cananda')
plt.show()
```

```python
col3=['India','Philiphines','China']
new=pd.DataFrame()
for i,j in zip([0,1,3],col3):
    new[j]=df.iloc[i,4:-1]
new1=new.T
#print(new)
ig=plt.figure(figsize=(20,8))
plt.boxplot(new1)
plt.show()
```

```python
new4=pd.DataFrame()
for i ,j in zip([10,11],['India','France']):
    new4[j]=df.iloc[i,4:-1]
    new4.plot.area()
```

```python
fig1,ax=plt.subplots()
fnew4=pd.DataFrame()
for i,j in zip([0,11],['India','France']):
    fnew4[j]=df.iloc[i,4:]
l4=['India','France']
pnew=fnew4.T
ax.pie(pnew['Total'],labels=l4)
plt.show()
```
